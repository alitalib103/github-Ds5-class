### 5. Solve following question by using qcut.

#### a. Enrollment Quantile Categories: How are trials distributed across four quantiles based on enrollment numbers?

```python
df['Enrollment Quantile']=pd.qcut(df["Enrollment"],q=4)
print(df[['Enrollment','Enrollment Quantile']])
```

```python
df['Enrollment Quantile'].unique()
```

```python
df['Enrollment Quantile'].value_counts()
```

#### b. Start Date Year Quantiles: What are the time periods represented by four quantiles of trial start years?

```python
df['Start Year'] = df['Start Date'].dt.year
df['Start Year'].value_counts()
```

```python
df['Start Date Year Quantiles'] = pd.qcut(df['Start Year'], q=4, duplicates='drop')
print(df[['Start Year','Start Date Year Quantiles']])
```

```python
df['Start Date Year Quantiles'].unique()
```

```python
df['Start Date Year Quantiles'].value_counts()
```

#### c. Completion Time Frame Categories: How do trial durations distribute across four quantiles from start to completion date?

```python
df['Start Date'] = pd.to_datetime(df['Start Date'], errors ='coerce')
df['Completion Date'] = pd.to_datetime(df['Completion Date'], errors ='coerce')
```

```python
df['Trial Duration'] = (df['Completion Date'] - df['Start Date']).dt.days
```

```python
df['Completion Time Frame']= pd.qcut(df["Trial Duration"],q=4)
print(df[['Trial Duration','Completion Time Frame']])
```

```python
df.groupby('Completion Time Frame')['Trial Duration'].agg(['min', 'max'])
```

```python
df['Completion Time Frame'].unique()
```

```python
df['Completion Time Frame'].value_counts()
```

#### d. Outcome Measures Count Quantiles: How are trials categorized into four quantiles based on the number of outcome measures?

```python
df['Outcome Measures Count'] = df['Outcome Measures'].str.split('|').str.len()
```

```python
df['Outcome Measures Count Quantiles'] = pd.qcut(df['Outcome Measures Count'], q=4)
print(df[['Outcome Measures Count','Outcome Measures Count Quantiles']])
```

```python
df.groupby('Outcome Measures Count Quantiles')['Outcome Measures Count'].agg(['min', 'max'])
```

```python
df['Outcome Measures Count Quantiles'].value_counts()
```

#### e. Trial Update Frequency Quantiles: What does the distribution of trial update frequencies look like across four quantiles?

```python
df['Last Update Posted'] = pd.to_datetime(df['Last Update Posted'], errors ='coerce')
```

```python
df['Trial Update Frequency'] = pd.qcut(df['Last Update Posted'], q=4)
print(df[['Last Update Posted','Trial Update Frequency']])
```

```python
df['Trial Update Frequency'].value_counts()
```

### 6. Completion Time Frame Grouping: Apply the cut function to categorize the completion time frames of trials into distinct intervals. What are the characteristics of trials within each time frame group?

```python
df.groupby('Completion Time Frame')['Trial Duration'].agg(['min', 'max'])
```

```python
intervals = [0, 365, 730, 1095, float('inf')]
labels = ['Less than 1 year', '1-2 years', '2-3 years', 'More than 3 years']
df['Completion Time Frame Group'] = pd.cut(df['Trial Duration'], bins=intervals, labels=labels, right=False)
```

```python
df[['Trial Duration','Completion Time Frame Group']]
```

```python
df['Completion Time Frame Group'].value_counts()
```

### 7. Enrollment Distribution Analysis: Use Matplotlib&#39;s hist() function to create a histogram visualizing the distribution of the &#39;Enrollment&#39; column in the dataset.

```python
df[['Enrollment']]
```

```python
plt.hist(df['Enrollment'])
plt.show() 
```

```python
# none- 0
# Low- 1-100
# Mid- 100-2000
# High- above 2000
intervals = [0, 100, 1999, 2000, float('inf')]
labels = ['None', 'Low', 'Mid', 'High']
df['Enrollment categories'] = pd.cut(df['Enrollment'], bins=intervals, labels=labels, right=False)
```

```python
df['Enrollment categories'].dtypes
```

```python
df['Enrollment categories'].value_counts()
```

```python
df['Enrollment categories'] = df['Enrollment categories'].astype(str)

plt.hist(df['Enrollment categories'], bins=10)
plt.xlabel('Enrollment categories')
plt.ylabel('Frequency')
plt.title('Enrollment Categories Distribution')
plt.show()
```
